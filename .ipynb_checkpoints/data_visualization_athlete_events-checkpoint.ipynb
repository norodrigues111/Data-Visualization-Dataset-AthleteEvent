### Analysing Different Scenarios and Generating the Appropriate Visualization 

The goal is to identify the top five sports based on the largest number of medals awarded 
in the year 2016, and then perform the following analysis: 
    
    1. Generate a plot indicating the number of medals awarded in each of the top five sports in 2016. 

    2. Plot a graph depicting the distribution of the age of medal winners in the top five sports in 2016.

    3. Find out which national teams won the largest number of medals in the top five sports in 2016. 

    4. Observe the trend in the average weight of male and female athletes winning in the top five sports in 2016. 



```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
relative_path = "../data/athlete_events.csv"
```

```python
df = pd.read_csv(relative_path)
```

```python
df.shape
```

```python
#Step 1 - Download the dataset and format it as a pandas DataFrame. 
df.head()
```

```python
#Step 2 - Filter the DataFrame to only include the rows corresponding to medal winners from 2016
winners_2016 = df[(df['Year']==2016) & (~df['Medal'].isna())]
winners_2016.head()
```

```python
#Step 3 - Find out the medals awarded in 2016 for each sport. 
medals_per_sport = winners_2016.groupby('Sport')['Medal'].count().reset_index()
medals_per_sport
```

```python
#Step 4:
#List the top five sports based on the largest number of medals awarded. 
#Filter the DataFrame one more time to only include the records for the top five sports in 2016
top_5_sports = medals_per_sport.sort_values(by='Medal',ascending=False).head()
top_5_sports
```

1. Generate a plot indicating the number of medals awarded in each of the top five sports 
in 2016.

```python
# Step 5 - Generate a bar plot of record counts corresponding to each of the top five sports. 
plt.figure(figsize=(6,4))
sns.barplot(x=top_5_sports['Sport'],y=top_5_sports['Medal'],palette='tab10')
plt.title('Number of medals of the TOP 5 sports in 2016')
plt.xlabel('Sport')
plt.ylabel('Count');
```

2. Plot a graph depicting the distribution of the age of medal winners in the top five 
sports in 2016.

```python
#Step 6 - Generate a histogram for the Age feature of all medal winners in the top five sports 2016
top_sports_list = top_5_sports['Sport'].tolist()
medal_winners_top_sports = winners_2016[winners_2016['Sport'].isin(top_sports_list)]
```

```python
plt.figure(figsize=(6,4))
plt.hist(medal_winners_top_sports['Age'],color='Skyblue')
plt.title('Number of medals of the TOP 5 sports in 2016')
plt.xlabel('Age');
```

3. Find out which national teams won the largest number of medals in the top five sports 
in 2016.

```python
medal_winners_per_country_2016 = medal_winners_top_sports.groupby('Team')['Medal'].count().reset_index()
medal_winners_per_country_2016 = medal_winners_per_country_2016.sort_values(by='Medal', ascending=False)
medal_winners_per_country_2016
```

```python
#Step 7 - Generate a bar plot indicating how many medals were won by each country's team in the top five sports in 2016. 
plt.figure(figsize=(10,6))
sns.barplot(data=medal_winners_per_country_2016, x=medal_winners_per_country_2016['Team'],y=medal_winners_per_country_2016['Medal'])
plt.title('Number of medals per country of the TOP 5 sports in 2016')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout();
```

4. Observe the trend in the average weight of male and female athletes winning in the 
top five sports in 2016. 

```python
#Step 8 - Generate a bar plot indicating the average weight of players, categorized based on gender, winning in the top five sports in 2016. 

sns.barplot(data=medal_winners_top_sports, x=medal_winners_top_sports['Sport'],y=medal_winners_top_sports['Weight'],hue='Sex')
plt.title('Average weight of winners of the TOP 5 sports in 2016')
plt.xlabel('Sport')
plt.ylabel('weight')
plt.tight_layout()
plt.legend(loc='upper right', bbox_to_anchor=(1.2,1));
```
